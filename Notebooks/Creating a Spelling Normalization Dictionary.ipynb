# Creating a list of Unique Tokens for Normalization

In this notebook, we take a sample of our texts and create a list of unique tokens that we will use to create a dictionary of normalized spellings. We also include a count of the instances of each token.

```python
#import modules we will use
from spacy.lang.es import Spanish 
import spacy
from spacy.tokens import Doc
from collections import defaultdict, Counter
from spacy.attrs import ORTH
import pandas as pd
import json
```

```python
# Load the model
nlp = spacy.load('es_core_news_md')
```

```python
#Read the documents

df = pd.read_csv('DocumentosTest.txt', delimiter =",")
print(df)
```

```python
# We join all the documents, treating them as one large text that can be tokenized 

docblock = " ".join(df.document)
print(docblock)
```

```python
#We pass our grouped document through the pipeline
docs= nlp(docblock)
```

```python
# And we print each token with its count 
counts = docs.count_by(ORTH)
for word_id, count in sorted(counts.items(), reverse=True, key=lambda item: item[1]):
    print(nlp.vocab.strings[word_id],",", count)
```
